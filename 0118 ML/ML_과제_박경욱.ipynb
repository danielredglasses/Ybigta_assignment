# 과제1

* 먼저, 지도 학습은 레이블이 있는 데이터를 학습하는 것으로 레이블이 numerical이라면 설명변수들의 값을 통해 이 레이블을 predict하는 것과 레이블이 categorical이라면 설명변수의 값을 통해 classification하는 task가 대표적인 예시이다. 반면, 비지도 학습은 레이블이 없는 데이터를 학습하는 것으로 데이터의 nearness를 통해 unlabeled 데이터를 clustering하는 것과 고차원 데이터를 저차원 데이터로 축소시키는 Dimensionality Reduction이 대표적인 예시이다.

* Clustering의 가장 대표적인 알고리즘으로는 K-means clustering이다. 이름 그대로 unlabaled data를 k개의 그룹으로 나누는 알고리즘이다. 알고리즘 방식은 다음과 같다. 첫 번째, unlabaled data들 중 랜덤으로 k개를 고르고, 이들을 cluster의 center, 즉 centroid라 부른다. 두 번째, 모든 data에 대해서 가장 가까운 centroid를 찾는다. 세 번째, 이렇게 생신 새로운 cluster들에 대해서 cluster 내의 data의 평균으로 새로운 centroid를 계산한다. 네 번째, 이를 cluster의 변화가 없을 때까지 계속 반복한다. k-means clustering의 단점으로는 맨 처음에 centroids를 랜덤으로 뽑을 때 어떤 게 뽑혔느냐에 따라서 결괏값이 다르게 나온다는 것이다. K-means clustering에서 nearness를 Euclidean distance가 아닌 다른 지표로 사용하는 방법도 존재하며, 다른 알고리즘으로는 Fuzzy K-means Algorithm, Hierarchical Clustering Algorithm, EM Algorithm 등이 있다.

* Dimensionality Reduction의 대표적인 예로는 PCA (Principal Component Analysis)이다. 데이터를 feature scaling과 mean normalization으로 전처리를 한 후 Σ를 계산하여 svd를 통해 component를 계산할 수 있다.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
# Normalize the data
df.iloc[:, :-1] = df.iloc[:, :-1].apply(lambda x : (x - x.mean()) / x.std(), axis = 0)
df
```

Classification

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

df_clf = DecisionTreeClassifier(random_state = 111)

X_train, X_test, y_train, y_test = train_test_split(df.drop('Outcome', axis = 1),
                                                    df['Outcome'],
                                                    test_size = 0.3, random_state = 111)

df_clf = df_clf.fit(X_train, y_train)
```

Evaluation

```python
# 평가지표 출력
from sklearn.metrics import accuracy_score
pred = df_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

rf_reg = RandomForestRegressor(n_estimators=100)
gbm_reg = GradientBoostingRegressor(n_estimators=100)
xgb_reg = XGBRegressor(n_estimators=100)
lgbm_reg = LGBMRegressor(n_estimators=100)
```

```python
model = rf_reg

from sklearn.metrics import mean_squared_error
def get_model_predict(model, X_train, X_test, y_train, y_test, is_expm1=False):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    if is_expm1:
        y_test = np.expm1(y_test)
        pred = np.expm1(pred)

    mse = mean_squared_error(y_test, pred)

    print('###', model.__class__.__name__, '###')
    print('Mean Squared Error:', mse)
```

```python
get_model_predict(model,X_train, X_test, y_train, y_test, is_expm1=True)
```

```python
def get_two_models_predict(model1, model2, X_train, X_test, y_train, y_test, is_expm1=False):
    model1.fit(X_train, y_train)
    model1_pred = model1.predict(X_test)
    model2.fit(X_train, y_train)
    model2_pred = model2.predict(X_test)

    pred = 0.5*model1_pred + 0.5*model2_pred

    if is_expm1 :
        y_test = np.expm1(y_test)
        pred = np.expm1(pred)
    mse = mean_squared_error(y_test, pred)

    print('### Two Models ###')
    print('Mean Squared Error:', mse)
```

```python
model1 = rf_reg
model2 = gbm_reg

get_two_models_predict(model1, model2, X_train, X_test, y_train, y_test, is_expm1=True)
```
