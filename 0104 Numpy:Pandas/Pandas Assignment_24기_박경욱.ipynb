### 151. Sort DataFrame based on another list

```python
df = pd.DataFrame([["A", 1], ["B", 2],
                   ["C", 3], ["D", 4]], columns=["col1", "col2"])

sort_list = ["C", "A", "D", "B"]

# start your code below
sort_dict = {}
for i in range(len(sort_list)):
    sort_dict[sort_list[i]] = i+1

new_df = df.sort_values(by = 'col1', key = lambda x : x.map(sort_dict))

# end your code here
print(new_df)
```
